```python
# QUES.1 What is the difference between a t-test and a z-test? Provide an example scenario where you would
# use each type of test.
# ANSWER  A t-test and a z-test are both statistical hypothesis tests used to assess whether the means of two groups are
# significantly different from each other. However, they differ in their assumptions and the situations in which they are
# typically applied.

# 1. T-test:
# * Assumption: The t-test assumes that the sample data follow a normal distribution and that the population standard
# deviation is unknown.
# * Use case: The t-test is commonly used when dealing with small sample sizes (typically fewer than 30 observations) or when 
# the population standard deviation is unknown. It is often used in medical research, social sciences, and other fields where
# sample sizes tend to be small.
# * Example scenario: Suppose you want to compare the mean scores of two groups of students (e.g., Group A and Group B) on a 
# standardized test. You would use a t-test to determine if there is a statistically significant difference between the mean
# scores of the two groups.
# 2. Z-test:
# * Assumption: The z-test assumes that the sample data follow a normal distribution and that the population standard 
# deviation is known.
# * Use case: The z-test is appropriate when dealing with large sample sizes (typically more than 30 observations) and when
# the population standard deviation is known. It is commonly used in quality control, manufacturing, and other fields where
# large datasets are available.
# * Example scenario: Imagine a company wants to test whether the mean weight of their product packaging matches the specified
# weight (which is known). They collect a large sample of packages and measure their weights. In this case, a z-test would
# be suitable to determine if the mean weight of the packages significantly differs from the specified value.
# In summary, the choice between a t-test and a z-test depends on factors such as sample size, knowledge of the population
# standard deviation, and the distribution of the data.
```

```python
# QUES.2 Differentiate between one-tailed and two-tailed tests.
# ANSWER One-tailed and two-tailed tests are two types of statistical hypothesis tests used to determine whether there is
# enough evidence to reject the null hypothesis in favor of the alternative hypothesis. The key difference between them lies
# in how they assess the significance of the relationship or difference being studied.

# 1. One-tailed test:
# * In a one-tailed test, the critical region (the area where the null hypothesis is rejected) is located entirely in one tail
# of the distribution curve (either the left or the right).
# * One-tailed tests are typically used when the researcher has a specific directional hypothesis about the relationship 
# between variables. For example, "A is greater than B" or "X has a positive effect on Y."
# * The decision to reject the null hypothesis is made based on whether the test statistic falls into the critical region in 
# that specified direction.
# * One-tailed tests are more powerful (i.e., have a higher chance of detecting an effect)when the direction of the effect is 
# known in advance.
# 2. Two-tailed test:
# * In a two-tailed test, the critical region is split between both tails of the distribution curve.
# * Two-tailed tests are used when the researcher does not have a specific directional hypothesis, or when they want to test
# for the possibility of both positive and negative effects.
# * The decision to reject the null hypothesis is based on whether the absolute value of the test statistic falls into either
# tail of the distribution.
# * Two-tailed tests are generally more conservative because they require a higher level of significance to reject the null
# hypothesis, as they are testing for the possibility of effects in both directions.
# In summary, the choice between one-tailed and two-tailed tests depends on the researcher's hypothesis and the
# directionality of the effect they are investigating. One-tailed tests are suitable when there is a specific directional
# hypothesis, while two-tailed tests are more appropriate when the direction of the effect is uncertain or when testing for
# the possibility of effects in both directions.
```

```python
# QUES.3 Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
# each type of error.
# ANSWER In hypothesis testing, Type 1 and Type 2 errors are two potential mistakes that can occur when making a decision
# about a hypothesis.

# 1. Type 1 Error: This error occurs when the null hypothesis (H0) is incorrectly rejected when it is actually true. In other 
# words, it's a false positive - you conclude that there is an effect or relationship when there isn't one. Type 1 errors 
# are also known as alpha errors or false alarms.Example Scenario: Imagine a medical test for a disease where the null 
# hypothesis is "the patient does not have the disease." A Type 1 error would occur if the test indicates that the patient 
# has the disease when they actually don't. This could lead to unnecessary treatments or stress for the patient.
# 2. Type 2 Error: This error occurs when the null hypothesis is incorrectly accepted when it is actually false. In other
# words, it's a false negative - you fail to detect an effect or relationship when there actually is one. Type 2 errors are
# also known as beta errors or misses.Example Scenario: Consider a clinical trial for a new drug where the null hypothesis 
# is "the drug has no effect." A Type 2 error would occur if the trial fails to detect that the drug is effective when it
# actually is. This could result in the drug not being approved for use, denying potentially beneficial treatment to patients.

# Both types of errors are important to consider in hypothesis testing because they impact the validity and reliability
# of the conclusions drawn from the analysis. The balance between Type 1 and Type 2 errors is often managed by selecting
# appropriate significance levels and sample sizes in statistical analyses.
```

```python
# QUES.4 Explain Bayes's theorem with an example.
# ANSWER Bayes's theorem is a fundamental concept in probability theory that describes how to update the probability of a
# hypothesis in light of new evidence. It's named after the Reverend Thomas Bayes, an 18th-century statistician and 
# the ologian.

# The theorem is expressed mathematically as:
# P(H∣E) = P(E∣H)×P(H)/P(E)

# Where:
# P(H∣E) is the probability of hypothesis H given evidence E (posterior probability).
# P(E∣H) is the probability of evidence E given hypothesis H (likelihood).
# P(H) is the prior probability of hypothesis H.
# P(E) is the prior probability of evidence E.
# Let's illustrate this with a classic example, often called the "diagnostic test" example:
# Suppose you have a medical test to detect a disease, and let's say the disease is quite rare, occurring in only 1% of the
# population (prior probability P(H)). The test is quite accurate, with a 95% chance of correctly detecting the disease if 
# it's present (likelihood  P(E∣H)), and a 2% chance of giving a false positive (incorrectly indicating the disease when 
# it's not there).
# Now, imagine you take the test and it comes back positive (evidence E). You want to know: What's the probability that you
# actually have the disease (posterior probability P(H∣E))

# Using Bayes's theorem:
# P(H∣E)= P(E∣H)×P(H)/P(E)
# Given:

# P(E∣H)=0.95 (the probability of a positive test given you have the disease)

# P(H)=0.01 (the prior probability of having the disease)

# P(E) can be calculated using the law of total probability:

# P(E)=P(E∣H)×P(H)+P(E∣¬H)×P(¬H)
# where ¬H represents not having the disease.
# Substituting the given values:
# P(E)=(0.95×0.01)+(0.02×0.99)=0.0293

# Now, putting it all together:
# P(H∣E)=0.95×0.01/0.0293≈0.324
# So, even though the test came back positive, there's only about a 32.4% chance that you actually have the disease. This 
# example demonstrates how Bayes's theorem allows us to update our beliefs (prior probabilities) based on new evidence.
```

```python
# QUES.5 What is a confidence interval? How to calculate the confidence interval, explain with an example.
# ANSWER A confidence interval is a range of values that likely contains the true value of a population parameter, such as
# a population mean or proportion, with a certain level of confidence. It quantifies the uncertainty associated with 
# estimating population parameters from sample data.

# Here's how you can calculate a confidence interval for the population mean with a known population standard deviation (σ):
# * Determine the sample mean (x̄) from your data.
# * Choose a confidence level, typically expressed as a percentage, such as 95%.
# * Look up the corresponding critical value from the standard normal distribution table or use a statistical software.
# Calculate the margin of error, which is the critical value multiplied by the standard deviation of the sample mean (σ/√n)
# , where n is the sample size.
# Construct the confidence interval by adding and subtracting the margin of error from the sample mean.
# Here's an example:

# Suppose we want to estimate the average height of adult males in a city. We take a random sample of 100 adult males and 
# find that the sample mean height is 175 cm, and the population standard deviation is known to be 8 cm. We want to construct
# a 95% confidence interval for the population mean height.

# Sample mean (x) = 175 cm
# Population standard deviation (σ) = 8 cm
# Sample size (n) = 100
# Confidence level = 95%
# Using a standard normal distribution table or a statistical software, the critical value for a 95% confidence interval is 
# approximately 1.96.
# Margin of error = Critical value × Standard deviation of sample mean
# =1.96×8/√100 = 1.96×0.8=1.568
# Confidence interval = Sample mean ± Margin of error
# =175±1.568
# So, the 95% confidence interval for the population mean height is approximately 175±1.568 or (173.432, 176.568) cm.
# This means we are 95% confident that the true average height of adult males in the city falls within the range of 173.432
# cm to 176.568 cm.
```

```python
# QUES.6 Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's
# probability and new evidence. Provide a sample problem and solution.
# ANSWER Suppose a factory produces light bulbs, and two machines, Machine A and Machine B, are used in the production
# process. Machine A produces 60% of the bulbs, while Machine B produces the remaining 40%. It is known that Machine A 
# produces 2% defective bulbs, whereas Machine B produces 5% defective bulbs. If a randomly selected bulb is defective,
# what is the probability that it came from Machine A?

# Solution:
# Let's define the events as follows:

# Event A: The bulb is produced by Machine A.
# Event B: The bulb is defective.
# We want to find P(A|B), the probability that the bulb came from Machine A given that it is defective.

# According to Bayes' Theorem:
# P(A∣B)=P(B∣A)×P(A)/P(B)

# We already know:

# P(A)=0.6 (the probability that the bulb comes from Machine A)

# P(B∣A)=0.02 (the probability that a bulb is defective given that it comes from Machine A)

# P(B∣B)=0.05 (the probability that a bulb is defective given that it comes from Machine B)

# To calculate 

# P(B), we can use the law of total probability:

# P(B)=P(B∣A)×P(A)+P(B∣B)×P(B)

# P(B)=0.02×0.6+0.05×0.4

# P(B)=0.012+0.02

# P(B)=0.032

# Now we can plug in the values into Bayes' Theorem: 
# P(A∣B)≈0.375
```

```python
# QUES.8 What is the margin of error in a confidence interval? How does sample size affect the margin of error?
# ANSWER Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
# ANSWER The margin of error (MOE) in a confidence interval is a measure of the uncertainty or precision of an estimate
# based on a sample. It represents the range within which the true population parameter is likely to lie. The margin of error
# is influenced by several factors, including the confidence level chosen, the variability of the data, and most notably, 
# the sample size.

# In general, as the sample size increases, the margin of error decreases. This is because larger sample sizes provide more 
# information about the population, resulting in a more accurate estimate of the population parameter.

# For example, let's consider a scenario where we want to estimate the average height of adults in a city. We take two
# samples: one with 50 individuals and another with 500 individuals.

# For the sample of 50 individuals, the margin of error might be relatively large because the sample size is small,
# resulting in greater uncertainty about how well it represents the entire population. However, for the sample of 
# 500 individuals, the margin of error would likely be smaller because the larger sample size allows for a more precise 
# estimation of the population parameter.

# So, in summary, larger sample sizes lead to smaller margins of error because they provide more reliable estimates of the
# population parameter due to increased precision and reduced sampling variability.
```

```python
# QUES.9 Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
# standard deviation of 5. Interpret the results.
# ANSWER To calculate the z-score, you can use the formula:
# Z=X−μ/σ
# Where:

# X is the data point value (in this case, 75).
# μ is the population mean (in this case, 70).
# σ is the population standard deviation (in this case, 5).
# Plugging in the values:


# Z= 75−70/5=5/5=1
# So, the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5 is 1.

# Interpretation: A z-score of 1 means that the data point is 1 standard deviation above the mean. In this case, the value 
# of 75 is one standard deviation above the population mean of 70.
```

```python
# QUES.10 In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
# of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
# significantly effective at a 95% confidence level using a t-test.
# ANSWER
# To conduct a hypothesis test to determine if the weight loss drug is significantly effective, we can use a t-test for the 
# mean weight loss. Here's how we proceed:

# Step 1: Define the null and alternative hypotheses:

# Null Hypothesis (H0): The weight loss drug is not significantly effective; the mean weight loss is equal to or less than
# zero.
# Alternative Hypothesis (H1): The weight loss drug is significantly effective; the mean weight loss is greater than zero.

# Step 2: Determine the significance level (α):

# Given a 95% confidence level, α = 0.05.

# Step 3: Calculate the test statistic:

# The formula for the t-test statistic for a sample mean is:
# t= x̄-μ/s/√n
# Where:
# x̄ is the sample mean (average weight loss),
# μ is the population mean under the null hypothesis (usually 0 in this case),
# s the population mean under the null hypothesis (usually 0 in this case),
# n is the sample size.
# Given: 
# x=6 pounds
# s=2.5 pounds
# n=50
# Let's calculate the t-test statistic.
# t≈16.97
# Step 4: Determine the critical value:

# Since this is a one-tailed test (we're testing if the drug is significantly effective in increasing weight loss), we need 
# to find the critical value for the upper tail of the t-distribution with degrees of freedom n−1.

# Using a t-table or statistical software with a significance level of 0.05 and degrees of freedom 
# df=49, we find the critical value to be approximately 1.676.

# Step 5: Make a decision:

# If the absolute value of the test statistic is greater than the critical value, we reject the null hypothesis.
# Here, 
# ∣16.97∣>1.676, so we reject the null hypothesis
# Step 6: Conclusion:

# Since we reject the null hypothesis, we have sufficient evidence to conclude that the weight loss drug is significantly 
# effective at a 95% confidence level.
```
